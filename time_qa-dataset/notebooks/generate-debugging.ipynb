```python
import json
from pathlib import Path

length = 3.0
target_dir = Path(".").absolute().parents[1] / "datasets-debug" / f"round-3-length_{length:.1f}-seed-3"
target_dir.mkdir(exist_ok=True, parents=True)
target_dir
```

```python
actions = """
playing guitar
punching
running
jumping once
drinking with the left hand
bowing
throwing a ball
shaking hands
skipping rope
picking something up with both hands
dancing
eating with the right hand
waving
catching a ball
kicking a ball
sitting down
golfing (swinging a club)
T-posing
holding a baby
"""

actions = actions.strip().lower().split("\n")

blocks = [actions[i : i + 5] for i in range(0, len(actions), 5)]

for idx, actions in enumerate(blocks):
    instance_path = target_dir / str(idx)
    instance_path.mkdir(exist_ok=True)

    meta_data = {
        "prompt_sequence": [
            {
                "start": index * length,
                "end": (index + 1) * length,
                "action": action,
            }
            for index, action in enumerate(actions)
        ],
    }

    with open(instance_path / "data.json", "w") as f:
        json.dump(meta_data, f, indent=4)
```
