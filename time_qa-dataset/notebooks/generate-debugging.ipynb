```python
import json
from pathlib import Path

length = 5.0
target_dir = Path(".").absolute().parents[1] / "datasets-debug" / f"length_{length:.1f}"
target_dir.mkdir(exist_ok=True, parents=True)
target_dir
```

```python
actions = """
playing guitar
punching
running
jumping once
drinking
Bowing
Playing a violin
throwing a ball with both hands
stretching
hugging
shaking hands
high-fiving
playing drums
skipping rope
sneezing
fishing (casting a line)
pushing a cart
doing a pirouette
Holding their own hand
Holding their head with both hands
Dancing
Writing on a board
Typing on a keyboard
Reading a book
Eating
eating with the right hand
Drinking
Brushing teeth
Combing hair
Clapping
Waving
Swimming (mimicking)
Catching a ball
Kicking a ball
Lifting weights
Playing piano
Knitting
Painting
Tying shoelaces
Stretching
Reading a newspaper
Texting on a phone
Cycling (mimicking)
Taking a selfie
Boxing (shadow boxing)
Washing hands
Jogging in place
Saluting
Surfing (mimicking)
Fishing (casting a line)
Golfing (swinging a club)
pointing downwards
putting on a jacket
T-posing
chopping
conducting
giving directions
grasping an object
hanging
holding a baby
"""

actions = actions.strip().lower().split("\n")

blocks = [actions[i : i + 5] for i in range(0, len(actions), 5)]

for idx, actions in enumerate(blocks):
    instance_path = target_dir / str(idx)
    instance_path.mkdir(exist_ok=True)

    meta_data = {
        "prompt_sequence": [
            {
                "start": index * length,
                "end": (index + 1) * length,
                "action": action,
            }
            for index, action in enumerate(actions)
        ],
    }

    with open(instance_path / "data.json", "w") as f:
        json.dump(meta_data, f, indent=4)
```
