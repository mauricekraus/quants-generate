## Setup

```python
import json
from pathlib import Path

import pandas as pd
```

```python
study_path = Path(".").absolute().parents[1] / "generated-dataset-30_000" / "human-study"

assert study_path.exists()
```

```python
original = pd.read_csv(study_path / "study_data_binary.csv")
# Remove qualifier and honeypot
original = original[original["study_entry_id"] >= 20]
original
```

```python
results = pd.read_csv(study_path / "results-binary-only-80.csv").rename(columns={"image_url": "url"})
results
```

```python
both = pd.merge(original, results, on="url")
# both = both[both["finished_on"].notnull()]

print(both[both["status"].isnull()])

both.drop(
    columns=[
        "created_on",
        "callback_metadata",
        "callback_url",
        "focus",
        "original_filename",
        "text_data",
        "finished_on",
    ],
    inplace=True,
)
both["status"] = both["status"].apply(lambda x: json.loads(x)[0])
both.set_index("study_entry_id", inplace=True)
both
```

```python
both[both["status"].isnull()]
```

```python
both["is_correct"] = both.apply(lambda x: x["answer"] == x["status"], axis=1)
good, bad = both.value_counts("is_correct")
print(f"Good: {good}, Bad: {bad}, Total: {good + bad}, Accuracy: {good / (good + bad) * 100:.2f}%")
```
